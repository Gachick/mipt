```python
import numpy as np 
import matplotlib.pyplot as plt
import pandas as pd
```

```python
C = 4.7 * 10 ** -8
L = 1.5 * 10 ** -3
nu_0= 1 / (np.pi * np.pow(L * C, .5))
N = 10
```

```python
#phi_nus = np.array([5.6, 11.1, 16.4, 21.4, 25.9, 29.4, 32.6, 34.8, 37]) * 1000
phi_nus = np.array([4, 8, 12, 16, 20, 24, 28, 32, 36, nu_0 / 1000]) * 1000
phi_theory = 2 * np.arcsin(phi_nus / nu_0)

phi_sins = np.array([5/11.5, 9/11, 11/11.5, 11.5/12, 8.5/12, 3/12, 3.5/13, 12/13, 12/14, 3.5/14.5])
# гадости из-за двойного угла, так как замеры через ячейку
phi_measured = np.arcsin(phi_sins) / 2
phi_measured[3:6] = np.pi / 2 - phi_measured[3:6]
phi_measured[6:8] = np.pi / 2 + phi_measured[6:8]
phi_measured[8:] = np.pi - phi_measured[8:]

plt.title(r"Зависимость $\varphi(\nu)$")
plt.plot(phi_nus, phi_theory, marker="+", ms=10, color="black", label="Теория")
plt.scatter(phi_nus, phi_measured, color="black", label="Фактические")
plt.xlabel(r"$\nu$, Гц")
plt.ylabel(r"$\varphi$, рад")
plt.legend()
```

```python
ms = np.array([i for i in range(N)])
z_inf_omegas = 2 / np.sqrt(L*C) * np.sin(ms * np.pi / (2 * N)) / (2 * np.pi)
z_0_omegas = 2 / np.sqrt(L*C) * np.sin(np.pi / (4 * N) + ms * np.pi / (2 * N)) / (2 * np.pi)
res = pd.DataFrame.from_records([z_inf_omegas, z_0_omegas], columns=ms)
res.index = ["z_inf", "z_0"]
res
```
